```python
# Extract year from date strings and filter
def extract_year(date_str):
    if pd.isna(date_str):
        return None
    try:
        # Extract first 4 digits as year (e.g., "1742-01-01T00:00:00Z" -> 1742)
        year = int(str(date_str)[:4])
        return year
    except:
        return None

# Apply year extraction
data['inception_year'] = data['inceptionDate'].apply(extract_year)
data['publication_year'] = data['publicationDate'].apply(extract_year)

# Filter: keep only rows where both years are <= 1800 or NaN
data = data[
    ((data['inception_year'] <= 1800) | (data['inception_year'].isna())) &
    ((data['publication_year'] <= 1800) | (data['publication_year'].isna()))
]

# Drop temporary columns if you want
data = data.drop(['inception_year', 'publication_year'], axis=1)
```

```python
import requests
import re
from urllib.parse import urlparse, unquote
import json
from tqdm import tqdm
from bs4 import BeautifulSoup

# Load language words
with open('wikisource_lang_words.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

VERSION_WORDS = data['version_words']
CATEGORY_WORDS = data['category_words']

def is_colored_background(style):
    """Check if an element has a colored/light background (not pure white)"""
    if not style:
        return False
    style = style.lower()
    
    if 'background' not in style and 'bgcolor' not in style:
        return False
    
    bg_match = re.search(r'background(?:-color)?:\s*([^;]+)', style)
    if not bg_match:
        return True
    
    bg_value = bg_match.group(1).strip()
    
    if bg_value in ['white', '#fff', '#ffffff', 'rgb(255,255,255)', 'rgb(255, 255, 255)']:
        return False
    
    if bg_value in ['transparent', 'none', 'inherit']:
        return False
    
    return True


def analyze_url(url):
    """
    Analyze a Wikisource URL and return its type:
    - 'direct_text': content is directly on the page
    - 'summary': has chapters (subpages or links without version mention)
    - 'versions': has links + mentions version/edition words in main content
    """
    parsed = urlparse(url)
    lang = parsed.netloc.split('.')[0]
    name = unquote(parsed.path.split('/wiki/')[-1])
    
    headers = {'User-Agent': 'MyBot/1.0 (myemail@example.com)'}
    
    try:
        response = requests.get(f"https://{lang}.wikisource.org/w/api.php",
            headers=headers,
            params={
                'action': 'parse',
                'page': name,
                'prop': 'text',
                'format': 'json'
            }
        ).json()
        
        if 'error' in response:
            return 'error', None
        
        html = response['parse']['text']['*']
        
        # Use BeautifulSoup
        soup = BeautifulSoup(html, 'html.parser')
        
        # Remove license containers
        for element in soup.find_all(class_=lambda c: c and any('license' in cls.lower() for cls in (c if isinstance(c, list) else [c]))):
            element.decompose()
        
        # Remove ALL elements with any background color (including light colors)
        for element in soup.find_all(style=is_colored_background):
            element.decompose()
        
        # Remove tables with bgcolor attribute
        for table in soup.find_all('table', bgcolor=True):
            table.decompose()
        
        # Remove tables with background in style
        for table in soup.find_all('table', style=is_colored_background):
            table.decompose()
        
        # Remove divs with background
        for div in soup.find_all('div', style=is_colored_background):
            div.decompose()
        
        # Remove td/th with background
        for cell in soup.find_all(['td', 'th'], style=is_colored_background):
            cell.decompose()
        for cell in soup.find_all(['td', 'th'], bgcolor=True):
            cell.decompose()
        
        # Remove category links by id
        for element in soup.find_all(id='catlinks'):
            element.decompose()
        
        # Remove footer elements
        for element in soup.find_all(id=lambda x: x and 'footer' in x.lower()):
            element.decompose()
        for element in soup.find_all(class_=lambda c: c and any('footer' in cls.lower() for cls in (c if isinstance(c, list) else [c]))):
            element.decompose()
        
        # Remove style and script tags
        for tag in soup.find_all(['style', 'script']):
            tag.decompose()
        
        # Clean text for word search
        text_clean = soup.get_text().lower()
        
        # Check for version words (case insensitive)
        version_words = VERSION_WORDS.get(lang, VERSION_WORDS['en'])
        has_version_word = any(word.lower() in text_clean for word in version_words)
        
        # Extract links from cleaned soup
        links = [a.get('href', '') for a in soup.find_all('a', href=True)]
        links = [l.replace('/wiki/', '') for l in links if l.startswith('/wiki/')]
        
        # Filter links
        filtered_links = []
        for link in links:
            link_decoded = unquote(link)
            if ':' in link_decoded:
                continue
            if link.startswith('#'):
                continue
            if link not in filtered_links:
                filtered_links.append(link)
        
        # Chapter links (subpages)
        chapter_links = [l for l in filtered_links if unquote(l).lower().startswith(name.lower() + '/')]
        
        # Other links
        other_links = [l for l in filtered_links if not unquote(l).lower().startswith(name.lower() + '/')]
        
        # Word count
        text = soup.get_text()
        text = re.sub(r'\s+', ' ', text).strip()
        word_count = len(text.split())
        
        # Total links count
        total_links = len(chapter_links) + len(other_links)
        
        # Decision logic
        if word_count > 50 and total_links < 3:
            page_type = 'direct_text'
        elif len(chapter_links) > 0:
            page_type = 'summary'
        elif has_version_word and len(other_links) > 3:
            page_type = 'versions'
        elif len(other_links) > 3 and word_count < 500:
            page_type = 'summary'
        else:
            page_type = 'direct_text'
        
        return page_type, {
            'url': url,
            'name': name,
            'lang': lang,
            'type': page_type,
            'words': word_count,
            'chapters': len(chapter_links),
            'other_links': len(other_links),
            'has_version_word': has_version_word,
            'text_clean': text_clean
        }
    
    except Exception as e:
        return 'error', None

# Test
url = "https://zh.wikisource.org/wiki/女冠子_(溫庭筠)"
page_type, data = analyze_url(url)

print(f"URL: {data['url']}")
print(f"Name: {data['name']}")
print(f"Lang: {data['lang']}")
print(f"Type: {data['type']}")
print(f"Words: {data['words']}")
print(f"Chapters: {data['chapters']}")
print(f"Other links: {data['other_links']}")
print(f"Has version word: {data['has_version_word']}")

# Check which version words were found
version_words = VERSION_WORDS.get(data['lang'], VERSION_WORDS['en'])
found_words = [word for word in version_words if word.lower() in data['text_clean']]
print(f"Found version words: {found_words}")
```

```python
import re

def extract_page_links(raw_text, base_lang='en'):
    if pd.isna(raw_text) or raw_text is None:
        return []
    
    # Find pattern like: <pages index="Book.djvu" from=16 to=18 />
    match = re.search(r'<pages\s+index="([^"]+)"\s+from=(\d+)\s+to=(\d+)', raw_text)
    
    if not match:
        return []
    
    index_name = match.group(1)
    from_page = int(match.group(2))
    to_page = int(match.group(3))
    
    # Build list of page URLs
    links = []
    for page_num in range(from_page, to_page + 1):
        url = f"https://{base_lang}.wikisource.org/wiki/Page:{index_name}/{page_num}"
        links.append(url)
    
    return links


# Example usage
raw = '<pages index="Sibylline Leaves (Coleridge).djvu" from=16 to=18 />'
print(extract_page_links(raw))
# ['https://en.wikisource.org/wiki/Page:Sibylline_Leaves_(Coleridge).djvu/16',
#  'https://en.wikisource.org/wiki/Page:Sibylline_Leaves_(Coleridge).djvu/17',
#  'https://en.wikisource.org/wiki/Page:Sibylline_Leaves_(Coleridge).djvu/18']

# Apply to your dataframe
df_grouped['page_links'] = df_grouped['raw_text'].apply(lambda x: extract_page_links(x, 'fr'))
```

```python
import plotly.express as px
import pandas as pd

# Remove duplicates first
df_clean = df_test.drop_duplicates(subset=['item', 'languageLabel']).copy()

# Combine Chinese and Classical Chinese
df_clean['languageLabel'] = df_clean['languageLabel'].replace({'Classical Chinese': 'Chinese'})

# Get top 15 languages
top_languages = df_clean['languageLabel'].value_counts().head(15).index

# Filter for top languages and valid years up to 1800
df_plot = df_clean[
    df_clean['languageLabel'].isin(top_languages) & 
    df_clean['year'].notna() & 
    (df_clean['year'] <= 1800)
].copy()

# Create century column - 1800 belongs to 1700s
df_plot['century'] = ((df_plot['year'] - 1) // 100) * 100

# Count unique items per language per century and reset index for plotly
plot_data = df_plot.groupby(['century', 'languageLabel'])['item'].nunique().reset_index()
plot_data.columns = ['century', 'Language', 'count']

# Create interactive plot
fig = px.line(
    plot_data, 
    x='century', 
    y='count', 
    color='Language',
    markers=True,
    log_y=False,
    title='Number of Unique Literary Works by Century (Top 15 Languages)',
    # labels={'century': 'Century', 'count': 'Number of Unique Items (log scale)'}
)

fig.update_layout(
    # hovermode='x unified',
    width=1000,
    height=600
)

fig.show()
```
